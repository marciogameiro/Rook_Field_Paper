## Example 1 (2D Example)

Define a DSGRN network

```python
net_spec = """x1 : (~x1)(~x2)
              x2 : (~x1)(~x2)"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

Construct the parameter graph

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Parameter graph size:', parameter_graph.size())
```

Choose a parameter index and get the corresponding parameter from the parametr graph

```python
# Parameter index
par_index = 974

parameter = parameter_graph.parameter(par_index)
```

Compute the multi-valued map and the Morse graph

```python
%%time
morse_graph, stg, graded_complex = DSGRN_utils.ConleyMorseGraph(parameter)
```

Plot the Morse graph

```python
DSGRN_utils.PlotMorseGraph(morse_graph)
```

Plot the cell complex and the multi-valued map

```python
DSGRN_utils.PlotMorseSets(morse_graph, stg, graded_complex)
```

## Example 16.5.1 (Non DSGRN ramp system)

To compute with a non DSGRN ramp system it is necessary to first define a ramp system. Let's define four ramp functions and then define the ramp nonlinearity of the ramp system (`f_ramp`).

```python
nu1_x = [7, 3, 7, 3]
theta1_x = [3, 5.1, 7.17]
h1_x = [0.15, 0.15, 0.15]
def fr1(x):
    return DSGRN_utils.ramp_function(x, nu1_x, theta1_x, h1_x)

nu1_y = [3.1, 4.9, 7.3]
theta1_y = [5.3, 6.1]
h1_y = [0.11, 0.11]
def gr1(y):
    return DSGRN_utils.ramp_function(y, nu1_y, theta1_y, h1_y)

nu2_x = [7.5, 5.5, 4]
theta2_x = [4, 5.5]
h2_x = [0.15, 0.15]
def fr2(x):
    return DSGRN_utils.ramp_function(x, nu2_x, theta2_x, h2_x)

nu2_y = [1.4, 5, 7.55, 9]
theta2_y = [4, 5.85, 7.55]
h2_y = [0.11, 0.11, 0.11]
def gr2(y):
    return DSGRN_utils.ramp_function(y, nu2_y, theta2_y, h2_y)

def f_ramp(x):
    return [fr1(x[0]) + gr1(x[1]), fr2(x[0]) + gr2(x[1])]
```

Define the parameters $\gamma$, $\theta$, and $h$ for the ramp system.

```python
gamma1 = 2.0
gamma2 = 2.0

gamma = [gamma1, gamma2]
theta = [[theta1_x, theta1_y], [theta2_x, theta2_y]]
h = [[h1_x, h1_y], [h2_x, h2_y]]
```

Compute the wall labelling for the ramp system and the number of thresholds in each dimension.

```python
labelling, num_thresholds = DSGRN_utils.ramp_system_wall_labelling(gamma, theta, h, f_ramp)
```

Compute the multi-valued map and the Morse graph from the wall labelling.

```python
%%time
morse_graph, stg, graded_complex = DSGRN_utils.ConleyMorseGraph(labelling=labelling, num_thresholds=num_thresholds)
```

```python
DSGRN_utils.PlotMorseGraph(morse_graph)
```

```python
DSGRN_utils.PlotMorseSets(morse_graph, stg, graded_complex, fig_w=10, fig_h=10)
```
